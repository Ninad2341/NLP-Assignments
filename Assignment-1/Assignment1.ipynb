```python
import nltk

# Download required datasets
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
text = "NLTK is amazing! I love machine learning, NLP, and deep-learning. Can't wait :) #AI"
```

```python
from nltk.tokenize import WhitespaceTokenizer

wt = WhitespaceTokenizer()
print(wt.tokenize(text))
```

```python
from nltk.tokenize import wordpunct_tokenize

print(wordpunct_tokenize(text))
```

```python
from nltk.tokenize import TreebankWordTokenizer

twt = TreebankWordTokenizer()
print(twt.tokenize(text))
```

```python
from nltk.tokenize import TweetTokenizer

tt = TweetTokenizer()
print(tt.tokenize(text))
```

```python
from nltk.tokenize import MWETokenizer

mwe = MWETokenizer([('machine', 'learning'), ('deep', 'learning')], separator='_')
print(mwe.tokenize(wordpunct_tokenize(text)))
```

```python
from nltk.stem import PorterStemmer

ps = PorterStemmer()
tokens = wordpunct_tokenize(text)
print([ps.stem(word) for word in tokens])
```

```python
from nltk.stem import SnowballStemmer

ss = SnowballStemmer("english")
print([ss.stem(word) for word in tokens])
```

```python
from nltk.stem import WordNetLemmatizer

wl = WordNetLemmatizer()
print([wl.lemmatize(word) for word in tokens])
```
